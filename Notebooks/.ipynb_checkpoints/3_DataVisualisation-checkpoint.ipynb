```python
import networkx as nx
import matplotlib.pyplot as plt
import math
```

```python
Graphe_Netflix = nx.read_edgelist("../Data/Liaisons.txt", delimiter = '\t', data=(('weight',float),))
```

```python
Graphe_Netflix.edges(data = True)
```

```python
Graphe_Netflix.nodes()
```

```python
plt.figure()
nx.draw_networkx(Graphe_Netflix)
plt.show()
```

```python
nx.draw(Graphe_Netflix,pos=nx.spring_layout(Graphe_Netflix))
```
