```python
%%time

init = [0]+[len(np.where(np.array(list(matrice.iloc[0,] + matrice.iloc[j,]))== 2)[0]) for j in range(1,Nb_films)]

Liste_poids = [init]

for i in range(1,Nb_films):
    liste_poids_ligne = [Liste_poids[j][i] for j in range(i)]+[0]
    for j in range(i+1, Nb_films):
        values = np.array(list(matrice.iloc[i,] + matrice.iloc[j,]))
        liste_poids_ligne.append(len(np.where(values == 2)[0]))
    Liste_poids.append(liste_poids_ligne)
```
